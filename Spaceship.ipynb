```python
ffill_cols = ['HomePlanet','CryoSleep','Cabin1','Cabin2','CabinNo.','Destination']
for col in ffill_cols:
    df[col]=df[col].fillna(method='ffill')
    test_df[col]=test_df[col].fillna(method='ffill')
```

```python
Agee = df[['Age']].values
Agee_test = test_df[['Age']].values
```

```python
from sklearn.impute import SimpleImputer
median_imputer=SimpleImputer(strategy='median')
median_imputer.fit(Agee)

df['Age']=median_imputer.transform(Agee)

median_imputer.fit(Agee_test)
test_df['Age']=median_imputer.transform(Agee_test)
```

```python
df['VIP'] = df['VIP'].fillna(df['TotalSpend'] > 10000)
test_df['VIP'] = test_df['VIP'].fillna(test_df['TotalSpend'] > 10000)
```

```python
constant_imputer = SimpleImputer(strategy='constant', fill_value=0)
spend_cols = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
df[spend_cols]= constant_imputer.fit_transform(df[spend_cols])
test_df[spend_cols]= constant_imputer.fit_transform(test_df[spend_cols])
```

```python
input_cols = [  'CryoSleep',  'Age',
       'VIP', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck',
         'TotalSpend', 'CabinNo.',
        'Destination_55 Cancri e', 'Destination_PSO J318.5-22',
       'Destination_TRAPPIST-1e', 'Cabin1_A', 'Cabin1_B', 'Cabin1_C',
       'Cabin1_D', 'Cabin1_E', 'Cabin1_F', 'Cabin1_G', 'Cabin1_T', 'Cabin2_P',
       'Cabin2_S', 'HomePlanet_Earth', 'HomePlanet_Europa', 'HomePlanet_Mars']
test_col = 'Transported'
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, shuffle=True,random_state=42)
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
```

```python
for train_index, val_index in kf.split(df):
    X_train, X_val = df.loc[train_index, input_cols], df.loc[val_index, input_cols]
    y_train, y_val = df.loc[train_index, test_col], df.loc[val_index, test_col]

    model = XGBClassifier(n_estimators=50, learning_rate=0.2, max_depth=5, random_state=42)
    model.fit(X_train,y_train)
    train_preds = model.predict(X_train)
    val_preds = model.predict(X_val)
    

    print(f"training:{accuracy_score(y_train,train_preds) }",f"val:{accuracy_score(y_val,val_preds)}")
    
```

```python
model = XGBClassifier(n_estimators=50, learning_rate=0.2, max_depth=5, random_state=42)
model.fit(df[input_cols],df[test_col])
test_pred = model.predict(test_df[input_cols])
test_pred
```

```python
sample_submission = pd.read_csv('sample_submission.csv')
sample_submission['Transported']=test_pred
sample_submission['Transported']=sample_submission['Transported'].astype(bool)
sample_submission.to_csv('submission.csv',index=False)
```
